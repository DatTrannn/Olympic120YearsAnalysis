```python
# import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# read dataset
athlete = pd.read_csv("athlete_events.csv")
region = pd.read_csv("noc_regions.csv")
athlete.head()
```

```python
region.head()
```

```python
# Join the dataframes

athlete_df = athlete.merge(region, how="left", on="NOC")
athlete_df.head()
```

```python
athlete_df.rename(columns={"region":"Region", "notes":"Notes"}, inplace=True)
```

```python
athlete_df.head()
```

```python
athlete_df.info()
```

```python
athlete_df.describe()
```

```python
# Check null values
athlete_df.isna().sum()
```

```python
# Vietnam details

athlete_df.query('Team == "Vietnam"').head()
```

```python
# Laos details

athlete_df.query('Team == "Laos"').head()
```

```python
# Top countries participating
top_10_countries = athlete_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# Plot for top 10 countries participants
plt.figure(figsize=(12,6))
sns.barplot(x=top_10_countries.index, y=top_10_countries)
plt.title("Overall Top 10 Participation by Country")
plt.ylabel("Number of Participants")
plt.xlabel("Countries")
```

```python
# Age distribution of participants

sns.histplot(data=athlete_df.Age, bins=np.arange(10, 80, 2), color="lightblue")
plt.title("Age Distribution of Athlets")
plt.ylabel("Number of Participants")
```

```python
# List winter sports

winter_sports = athlete_df[athlete.Season == "Winter"].Sport.unique()
winter_sports
```

```python
# List summer sports

summer_sports = athlete_df[athlete.Season == "Summer"].Sport.unique()
summer_sports
```

```python
# Male and Female Participants

gender_count = athlete_df.Sex.value_counts()
gender_count
```

```python
# Pie plot for Male and Female Participants
plt.title("Gender Distribution")
plt.pie(gender_count, labels=gender_count.index, autopct="%.1f%%")
```

```python
# Total medals

athlete_medals = athlete_df.Medal.value_counts()
athlete_medals
```

```python
# Total of Female Athletes in each Olympic

female_participants = athlete_df.query("Sex == 'F'")[["Sex","Year", "Season"]]
female_participants = female_participants.groupby(["Year", "Season"]).count().reset_index()
female_participants
```

```python
sns.catplot(x="Year", y="Sex", data=female_participants, kind="bar", hue="Season", height=5, aspect=3)
plt.title("Women Participants")
plt.ylabel("Count")
```

```python
# Gold medal athletes

goldMedals = athlete_df[(athlete_df.Medal == "Gold")]
goldMedals
```

```python
# Gold medal beyond 60
goldMedals['ID'][goldMedals['Age'] > 60].count()
```

```python
sporting_event = goldMedals["Sport"][goldMedals["Age"] > 60]
sporting_event
```

```python
# Plot for sporting_event

plt.figure(figsize=(10,5))
sns.countplot(sporting_event)
plt.title("Gold Medals for Athlete over 60 years")
```

```python
# Gold medals from each country

goldMedalsCountry = goldMedals.Region.value_counts().reset_index(name="Medal_num").head(5)
goldMedalsCountry
```

```python
# plot top 5 gold medal countries

sns.barplot(x="index", y = "Medal_num", data = goldMedalsCountry, palette="rocket")
plt.title("Gold Medals per Country")
plt.xlabel("Country")
plt.ylabel("Number of Medals")
```

```python
max_year = athlete_df.Year.max()

team_names = goldMedals[goldMedals.Year == max_year].Team
team_names.value_counts().head(10)
```

```python
sns.barplot(x=team_names.value_counts().head(20), y=team_names.value_counts().head(20).index)
plt.xlabel("Country's Medals for the year 2016")
```

```python
not_null_medals = athlete_df[(athlete_df.Height.notnull()) & (athlete_df.Weight.notnull())]
```

```python
sns.scatterplot(x="Height", y="Weight", data=not_null_medals, hue="Sex")
plt.title("Height vs Weight of Olympic Medalists")
```
